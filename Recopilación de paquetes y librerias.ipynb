## 1. Ramdom

```python
import random
```

### ¿Qué es el módulo random?

El módulo `random` es una librería estándar de Python que proporciona funciones para generar números y selecciones aleatorias. Está incluido en la instalación base de Python, por lo que no requiere instalación adicional mediante herramientas como pip. Este módulo es fundamental cuando necesitas introducir variabilidad controlada en tus programas.

### Funcionalidad Principal

Al importar `random` con `import random`, obtienes acceso a funciones clave como `random.choice()`, que selecciona un elemento aleatorio de una lista, y `random.randint()`, que genera números enteros aleatorios dentro de un rango especificado. Estas funciones son la base para crear comportamientos impredecibles en aplicaciones, desde juegos hasta simulaciones.

```python
# Ejemplos de uso del módulo random

import random

# Generar un número aleatorio entre 1 y 10
numero_aleatorio = random.randint(1, 10)
print(f"Número aleatorio entre 1 y 10: {numero_aleatorio}")

# Generar un número aleatorio entre 0 y 1
numero_aleatorio_float = random.random()
print(f"Número aleatorio entre 0 y 1: {numero_aleatorio_float}")

# Elegir un elemento aleatorio de una lista
colores = ['rojo', 'verde', 'azul', 'amarillo']
color_aleatorio = random.choice(colores)
print(f"Color aleatorio: {color_aleatorio}")
```

## 2. Requests

```python
import requests
```

`Requests`, una librería de Python especializada en realizar peticiones HTTP. Es como traer una caja de herramientas a tu código que te permite comunicarte con servidores web, APIs, extraer información de una página o enviar el contenido de un formulario de manera automatizada de forma sencilla.

### Conceptos clave

Petición GET: es la forma más común de solicitar datos a un servidor. Cuando escribes `requests.get('https://...')`, le estás diciendo al servidor **"dame la información de esta URL"**.

**Respuesta:** el servidor devuelve un objeto `response` que contiene los datos (`.json() `convierte JSON a diccionario, `.content` obtiene bytes, `.text` obtiene texto).

Para más información y tutoriales se recomienda leer: [J2LOGO](https://j2logo.com/python/python-requests-peticiones-http/)

```python
# Ejemplo de uso del módulo requests

import requests

# Hacer una solicitud GET a una API pública
response = requests.get('https://api.github.com')
if response.status_code == 200:
    print("Solicitud exitosa")
    data = response.json()
    print(f"Información del usuario de GitHub: {data['current_user_url']}")
else:
    print(f"Error en la solicitud: {response.status_code}")
```

## 3. Datetime

```python
import datetime
```

La instrucción `import datetime` trae el módulo datetime, de la biblioteca estándar de Python. Ese módulo define objetos como `date`, `time` y `datetime` que representan fechas y horas de forma estructurada, junto con funciones para crear, comparar, formatear y convertir entre ellos.

- Puedes obtener la fecha de hoy con `datetime.date.today()`.
- Para parsear una cadena en formato `"YYYY‑MM‑DD"` mediante` datetime.datetime.strptime(...)`. 
- Al trabajar con valores de fecha como objetos en lugar de simples `str`, puedes ordenarlos, restarlos para calcular diferencias, añadirle días, etc., sin preocuparte por el manejo manual de meses y años.

Documentación de `datetime`: [docs.python.org](https://docs.python.org/es/3.8/library/datetime.html)

Información extra: [tAprende con Alf.es](https://aprendeconalf.es/docencia/python/manual/datetime/)

```python
# Ejemplo de uso del módulo datetime

import datetime

# Obtener la fecha y hora actual
fecha_hora_actual = datetime.datetime.now()
print(f"Fecha y hora actual: {fecha_hora_actual}")

# Crear una fecha específica
fecha_especifica = datetime.datetime(2022, 1, 1)
print(f"Fecha específica: {fecha_especifica}")

# Calcular la diferencia entre dos fechas
diferencia = fecha_hora_actual - fecha_especifica
print(f"Diferencia entre fechas: {diferencia.days} días")
```
