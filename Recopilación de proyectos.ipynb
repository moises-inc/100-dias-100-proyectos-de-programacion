## Opci√≥n 2

Una opci√≥n mejorada, usando una API, cortecia de Gemini.

```python
# Generar un c√≥digo que seleccione aleatoriamente una comida y la imprima en pantalla la receta e informaci√≥n de la misma.

# Opci√≥n de API p√∫blica: TheMealDB (https://www.themealdb.com/api.php)
import requests

def obtener_receta_aleatoria():
    # URL de la API p√∫blica para una receta aleatoria
    url = "https://www.themealdb.com/api/json/v1/1/random.php"

    try:
        response = requests.get(url)
        response.raise_for_status() # Verificar si hubo errores en la petici√≥n
        data = response.json()

        # La API devuelve una lista llamada 'meals' con un solo elemento
        receta = data['meals'][0]

        # 1. Imprimir T√≠tulo y Categor√≠a
        print("="*60)
        print(f"üçΩÔ∏è  RECETA: {receta['strMeal']}")
        print(f"üåç  Cocina: {receta['strArea']} | Categor√≠a: {receta['strCategory']}")
        print("="*60)

        # 2. Imprimir Ingredientes
        # En esta API, los ingredientes y medidas est√°n en claves separadas (strIngredient1, strMeasure1, etc.)
        print("\nüõí INGREDIENTES:")
        for i in range(1, 21): # La API soporta hasta 20 ingredientes
            ingrediente = receta.get(f"strIngredient{i}")
            medida = receta.get(f"strMeasure{i}")

            # Si el ingrediente existe y no est√° vac√≠o, lo imprimimos
            if ingrediente and ingrediente.strip():
                print(f" - {medida.strip()} {ingrediente.strip()}")

        # 3. Imprimir Instrucciones
        print("\nüë®‚Äçüç≥ PASOS A SEGUIR:")
        instrucciones = receta['strInstructions']
        # Hacemos un poco de limpieza b√°sica del texto
        print(instrucciones.replace('. ', '.\n'))

        # 4. Informaci√≥n Extra
        if receta['strYoutube']:
            print(f"\nüì∫ Video Tutorial: {receta['strYoutube']}")
        
        if receta['strSource']:
            print(f"üîó Fuente original: {receta['strSource']}")

    except requests.exceptions.RequestException as e:
        print(f"‚ùå Error al conectar con la base de datos: {e}")

if __name__ == "__main__":
    obtener_receta_aleatoria()
```

# D√≠a 2 - Calendario de √°nimo

```python
# Generar un calendario de √°nimo, donde se registra como te vas sintiendo cada d√≠a y se guarda el resultado en un archivo de texto.

import datetime
import os
import json
from collections import defaultdict
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# Definir el archivo donde se guardar√°n los datos
DATA_FILE = "calendario_animo.txt"
# Funci√≥n para cargar los datos desde el archivo
def cargar_datos():
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, "r") as f:
            return json.load(f)
    return {}
# Funci√≥n para guardar los datos en el archivo
def guardar_datos(datos):
    with open(DATA_FILE, "w") as f:
        json.dump(datos, f)
# Funci√≥n para registrar el √°nimo del d√≠a
def registrar_animo(fecha, animo):
    datos = cargar_datos()
    datos[fecha] = animo
    guardar_datos(datos)
# Funci√≥n para graficar el √°nimo a lo largo del tiempo
def graficar_animo():
    datos = cargar_datos()
    fechas = []
    animos = []
    for fecha_str, animo in sorted(datos.items()):
        fecha = datetime.datetime.strptime(fecha_str, "%Y-%m-%d").date()
        fechas.append(fecha)
        animos.append(animo)
    plt.figure(figsize=(10, 5))
    plt.plot(fechas, animos, marker='o')
    plt.title("Calendario de √Ånimo")
    plt.xlabel("Fecha")
    plt.ylabel("√Ånimo")
    plt.ylim(0, 10)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
    plt.gcf().autofmt_xdate()
    plt.grid()
    plt.show()
# Men√∫ principal
def main():
    while True:
        print("1. Registrar √°nimo del d√≠a")
        print("2. Graficar √°nimo")
        print("3. Salir")
        opcion = input("Seleccione una opci√≥n: ")
        if opcion == "1":
            fecha = input("Ingrese la fecha (YYYY-MM-DD) o deje en blanco para hoy: ")
            if not fecha:
                fecha = datetime.date.today().strftime("%Y-%m-%d")
            animo = int(input("Ingrese su √°nimo (0-10): "))
            registrar_animo(fecha, animo)
            print("√Ånimo registrado.")
        elif opcion == "2":
            graficar_animo()
        elif opcion == "3":
            break
        else:
            print("Opci√≥n no v√°lida. Intente de nuevo.")
if __name__ == "__main__":
    main()
```
