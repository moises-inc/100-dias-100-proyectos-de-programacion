# DÃ­a 2 - Calendario de Ã¡nimo

```python
# Generar un calendario de Ã¡nimo, donde se registra como te vas sintiendo cada dÃ­a y se guarda el resultado en un archivo de texto.

import datetime
import os
import json
from collections import defaultdict
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# Definir el archivo donde se guardarÃ¡n los datos
DATA_FILE = "calendario_animo.txt"
# FunciÃ³n para cargar los datos desde el archivo
def cargar_datos():
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, "r") as f:
            return json.load(f)
    return {}
# FunciÃ³n para guardar los datos en el archivo
def guardar_datos(datos):
    with open(DATA_FILE, "w") as f:
        json.dump(datos, f)
# FunciÃ³n para registrar el Ã¡nimo del dÃ­a
def registrar_animo(fecha, animo):
    datos = cargar_datos()
    datos[fecha] = animo
    guardar_datos(datos)
# FunciÃ³n para graficar el Ã¡nimo a lo largo del tiempo
def graficar_animo():
    datos = cargar_datos()
    fechas = []
    animos = []
    for fecha_str, animo in sorted(datos.items()):
        fecha = datetime.datetime.strptime(fecha_str, "%Y-%m-%d").date()
        fechas.append(fecha)
        animos.append(animo)
    plt.figure(figsize=(10, 5))
    plt.plot(fechas, animos, marker='o')
    plt.title("Calendario de Ãnimo")
    plt.xlabel("Fecha")
    plt.ylabel("Ãnimo")
    plt.ylim(0, 10)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
    plt.gcf().autofmt_xdate()
    plt.grid()
    plt.show()
# MenÃº principal
def main():
    while True:
        print("1. Registrar Ã¡nimo del dÃ­a")
        print("2. Graficar Ã¡nimo")
        print("3. Salir")
        opcion = input("Seleccione una opciÃ³n: ")
        if opcion == "1":
            fecha = input("Ingrese la fecha (YYYY-MM-DD) o deje en blanco para hoy: ")
            if not fecha:
                fecha = datetime.date.today().strftime("%Y-%m-%d")
            animo = int(input("Ingrese su Ã¡nimo (0-10): "))
            registrar_animo(fecha, animo)
            print("Ãnimo registrado.")
        elif opcion == "2":
            graficar_animo()
        elif opcion == "3":
            break
        else:
            print("OpciÃ³n no vÃ¡lida. Intente de nuevo.")
if __name__ == "__main__":
    main()
```

# DÃ­a 3

Honestamente, no sabÃ­a que era un "Campo de Flujo de Ruido Doble", pero, luego de investigar, encontrÃ© queÂ Campo de Flujo de RuidoÂ Doble es: "EsÂ una tÃ©cnica de arte generativo utilizada para crear animaciones y patrones visuales que simulan fluidos, viento o movimientos orgÃ¡nicos."

Gracias a Gemini, obtenemos que:

1. Â¿QuÃ© es un Campo de Flujo (Flow Field)?

Imagina una cuadrÃ­cula invisible sobre el lienzo (la pantalla). En cada punto de esa cuadrÃ­cula hay una flecha (un vector) que apunta hacia una direcciÃ³n especÃ­fica.

Cuando lanzas "partÃ­culas" (puntos o lÃ­neas) sobre este lienzo, ellas no se mueven al azar, sino que siguen las direcciones de estas flechas, como si fueran hojas flotando en un rÃ­o con corrientes.

2. El papel del "Ruido" (Noise)

Si las flechas apuntaran a direcciones totalmente aleatorias, el movimiento serÃ­a caÃ³tico y feo. Para que se vea suave y orgÃ¡nico (como humo o agua), se utiliza Ruido de Perlin o Simplex Noise.

Este algoritmo genera nÃºmeros aleatorios que estÃ¡n "conectados" entre sÃ­; si un punto tiene un valor, el punto vecino tendrÃ¡ un valor muy similar. Esto hace que las flechas giren suavemente a travÃ©s de la pantalla.

3. Â¿Por quÃ© "Doble" Ruido?

En un campo de flujo bÃ¡sico, el ruido solo controla una variable: el Ãngulo (la direcciÃ³n de la flecha). Todas las partÃ­culas se mueven a la misma velocidad constante.

En un Campo de Flujo de Ruido Doble, se utilizan dos valores de ruido independientes (o dos capas de ruido) para controlar dos propiedades distintas de cada partÃ­cula simultÃ¡neamente. En el contexto del reto de Florin Pop (inspirado por el artista Alca), esto generalmente significa:

Ruido 1 (Ãngulo): Determina hacia dÃ³nde se mueve la partÃ­cula (la direcciÃ³n de la curva).

Ruido 2 (Velocidad/Longitud): Determina quÃ© tan rÃ¡pido se mueve o quÃ© tan larga es la lÃ­nea que dibuja en ese momento.

El resultado visual: Al combinar ambos, obtienes un efecto mucho mÃ¡s dinÃ¡mico. Algunas Ã¡reas del lienzo harÃ¡n que las partÃ­culas se muevan rÃ¡pido y recto, mientras que en otras se moverÃ¡n lento y girarÃ¡n mucho, creando esa textura compleja y "fibrosa" que ves en el proyecto.

Resumen tÃ©cnico del algoritmo:
- Se crea una cuadrÃ­cula de vectores.
- Para cada partÃ­cula, se calcula su posiciÃ³n $(x, y)$.
- Se obtiene un valor de ruido n1 para el Ã¡ngulo (usando noise $(x, y))$.
- Se obtiene un segundo valor de ruido n2 (usualmente desplazando las coordenadas, ej: noise $(x + 1000, y + 1000)$) para la velocidad o longitud.
- Se actualiza la posiciÃ³n de la partÃ­cula usando ambos valores.


```python
# Finalmente, un cÃ³digo en Python que cree un campo de flujo animado utilizando doble ruido para controlar la direcciÃ³n y velocidad de los puntos en pantalla.
# Cortecia de Gemini

import pygame
import random
import math
import noise 

# --- CONFIGURACIÃ“N INICIAL ---
WIDTH, HEIGHT = 800, 600
FPS = 60
NUM_PARTICLES = 1500  # Cantidad de puntos (ajÃºstalo segÃºn la potencia de tu PC)
BG_COLOR = (0, 0, 0)  # Fondo negro

# --- PARÃMETROS DEL DOBLE RUIDO ---
# Escala: quÃ© tan "zoomed in" estÃ¡ el mapa de ruido. Valores mÃ¡s bajos = curvas mÃ¡s suaves.
NOISE_SCALE = 0.005 
# Velocidad del tiempo: quÃ© tan rÃ¡pido cambia el patrÃ³n de flujo.
TIME_SPEED = 0.003
# Offset Z: La variable que cambia con el tiempo para animar el ruido.
z_off = 0
# Offset para separar las dos capas de ruido (Ã¡ngulo y velocidad)
SPEED_LAYER_OFFSET = 5000

# --- CLASE PARTÃCULA ---
class Particle:
    def __init__(self):
        # PosiciÃ³n inicial aleatoria en la pantalla
        self.pos = pygame.Vector2(random.randint(0, WIDTH), random.randint(0, HEIGHT))
        self.vel = pygame.Vector2(0, 0)
        self.acc = pygame.Vector2(0, 0)
        self.max_speed = random.uniform(1.5, 3.0) # Velocidad base variada
        # Color base (cyan/azulado)
        self.base_color = pygame.Color(0, 200, 255)

    def follow_field(self, z_time):
        # --- AQUÃ ESTÃ LA MAGIA DEL DOBLE RUIDO ---

        # RUIDO 1: ÃNGULO (DirecciÃ³n)
        # Usamos pnoise3 para ruido 3D (x, y, tiempo)
        angle_noise = noise.pnoise3(self.pos.x * NOISE_SCALE, 
                                    self.pos.y * NOISE_SCALE, 
                                    z_time)
        # Mapeamos el ruido (aprox -1 a 1) a un Ã¡ngulo en radianes (rotaciÃ³n completa)
        angle = angle_noise * math.pi * 4

        # RUIDO 2: VELOCIDAD (Magnitud)
        # Usamos un offset gigante en X e Y para leer una zona "diferente" del ruido
        speed_noise = noise.pnoise3((self.pos.x + SPEED_LAYER_OFFSET) * NOISE_SCALE, 
                                    (self.pos.y + SPEED_LAYER_OFFSET) * NOISE_SCALE, 
                                    z_time)
        # Normalizamos speed_noise para que sea un multiplicador positivo (ej. entre 0.2 y 1.8)
        speed_multiplier = (speed_noise + 1.2) * 0.8
        
        # Creamos el vector de direcciÃ³n basado en el Ã¡ngulo
        dir_vector = pygame.Vector2(math.cos(angle), math.sin(angle))
        
        # Aplicamos la fuerza: direcciÃ³n * velocidad variable
        force = dir_vector * speed_multiplier
        self.apply_force(force)

    def apply_force(self, force):
        self.acc += force

    def update(self):
        self.vel += self.acc
        # Limitamos la velocidad mÃ¡xima para que no se descontrolen
        if self.vel.length() > self.max_speed:
             self.vel.scale_to_length(self.max_speed)
        self.pos += self.vel
        self.acc *= 0 # Reseteamos la aceleraciÃ³n para el siguiente frame

        # Manejo de bordes: si sale por un lado, aparece por el opuesto
        if self.pos.x > WIDTH: self.pos.x = 0
        if self.pos.x < 0: self.pos.x = WIDTH
        if self.pos.y > HEIGHT: self.pos.y = 0
        if self.pos.y < 0: self.pos.y = HEIGHT

    def draw(self, surface):
        # Variamos ligeramente el brillo segÃºn la velocidad actual para efecto visual
        current_speed_ratio = self.vel.length() / self.max_speed
        color = self.base_color.lerp((255, 255, 255), current_speed_ratio * 0.5)
        # Dibujamos un pequeÃ±o cÃ­rculo
        pygame.draw.circle(surface, color, (int(self.pos.x), int(self.pos.y)), 1)

# --- SETUP DE PYGAME ---
pygame.init()
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Campo de Flujo de Doble Ruido Animado")
clock = pygame.time.Clock()

# Creamos la lista de partÃ­culas
particles = [Particle() for _ in range(NUM_PARTICLES)]

# Superficie para el efecto de rastro (fade effect)
trail_surface = pygame.Surface((WIDTH, HEIGHT), pygame.SRCALPHA)
# Rellenamos con negro semitransparente (el Ãºltimo valor es alpha: 0-255)
# Menor alpha = rastros mÃ¡s largos. Mayor alpha = rastros mÃ¡s cortos.
trail_surface.fill((0, 0, 0, 25)) 


# --- BUCLE PRINCIPAL ---
running = True
while running:
    clock.tick(FPS)

    # 1. Manejo de eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False

    # 2. LÃ³gica de actualizaciÃ³n
    # Avanzamos el tiempo para que el patrÃ³n de ruido cambie
    z_off += TIME_SPEED

    for p in particles:
        p.follow_field(z_off)
        p.update()

    # 3. Dibujado
    # En lugar de limpiar la pantalla, dibujamos la capa semitransparente
    # Esto es lo que crea las estelas.
    screen.blit(trail_surface, (0, 0))
    
    for p in particles:
        p.draw(screen)

    pygame.display.flip()

pygame.quit()
```
